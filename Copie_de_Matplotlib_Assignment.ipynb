```
#1.Import Matplotlib & Numpy
import numpy as np
from matplotlib import pyplot as plt
```

```
#2.Preparing data
x = np.arange(0,100)
y = np.random.randint(low = 0 , high = 1000 , size = 100)
print(len(x))
print(len(y))
```

```
#3.Draw a line chart
plt.plot(x,y)
plt.show()
```

```
#4.Config the axes
fig = plt.figure(figsize = (15,6))
plt.plot(x,y,label='line 1')
plt.plot()

plt.title("Line graph")
plt.xlabel('a-axis')
plt.ylabel('b-axis')
plt.legend()
plt.show()
```

```
#5.Multiple line chart
x = np.arange(100)
y1 = np.random.randint(low = 0 , high = 1000 , size = 100)
y2 = np.random.randint(low = 0 , high = 1000 , size = 100)
fig = plt.figure(figsize = (20,8))
plt.plot(x,y1,'ys-',label='line 1')
plt.plot(x,y2,'go--',label='line 2')
plt.title("Line Chart")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

```
#6.Draw a box plot
data_to_plot = [y1,y2]
fig=plt.figure(figsize=(20,10))
labels = ["Y1" , "Y2"]
plt.ylabel("Observed Values")
plt.title("Box plot")
plt.boxplot(data_to_plot)
plt.show()
```

```
#7.Draw a histogram
x = np.random.standard_normal(1000)
y = np.random.normal(1,1,1000)
plt.hist(x ,bins=20 ,color='b')
plt.hist(y ,bins=20 ,color='g')
plt.title("Histogram")
plt.show()
```

```
`#8.Draw a bar plot
objects = ('Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp')
y_pos = np.arange(len(objects))
performance = [10,8,6,4,2,1]

plt.bar(y_pos, performance, align='center', alpha=0.5)
plt.xticks(y_pos, objects)
plt.ylabel('Usage')
plt.title('Programming language usage')

plt.show()
```

```
#9.Draw a scatter plot
x = [50 ,10 ,20 ,46 ,50 ,45 ,200 ,100 ,50]
y1 = [20 ,100 ,40 ,56 ,41 ,46 ,200 ,80 ,60]
y2 = [60 ,50 ,100 ,200 ,50 ,40 ,60 ,200 ,40]
fig = plt.figure()
plt = fig.add_axes([0,0,1,1])
plt.scatter(x,y1, color='b')
plt.scatter(x,y2, color='r')
plt.set_xlabel('range')
plt.set_ylabel('score')
plt.set_title('plot')
plt.legend()
```

```
#10 Subplot
fig , a = plt.subplots(3,2)
a[0][0].plot(x,y1, label='line 1')
a[0][0].set_title("Line graph")
a[0][0].set_xlabel("a-axis")
a[0][0].set_ylabel("b-axis")
a[0][0].legend()
a[0][1].plot(x,y1,'ys-',label='line 1')
a[0][1].plot(x,y2,'go--',label='line 2')
a[0][1].set_xlabel("X-axis")
a[0][1].set_ylabel("Y-axis")
a[0][1].set_title("Line Chart")
a[0][1].legend()
a[1][0].labels = ["Y1" , "Y2"]
a[1][0].set_ylabel("Observed Values")
a[1][0].set_title("Box plot")
a[1][0].boxplot(data_to_plot)
a[1][1].hist(x ,bins=20 ,color='b')
a[1][1].hist(y ,bins=20 ,color='g')
a[1][1].set_title("Histogram")
a[2][0].bar(y_pos, performance, align='center', alpha=0.5)
a[2][0].set_ylabel('Usage')
a[2][0].set_title('Programming language usage')
a[2][1].scatter(x,y1, color='b')
a[2][1].scatter(x,y2, color='r')
a[2][1].set_xlabel('range')
a[2][1].set_ylabel('score')
a[2][1].set_title('plot')
a[2][1].legend()

plt.tight_layout()
plt.show()
```
